````python
# CELLA 4: Scarica l'APK
from google.colab import files
import glob

print("üîç Cerco l'APK compilato...")

# Cerca il file APK
apk_files = glob.glob('bin/*.apk')

if apk_files:
    apk_path = apk_files[0]
    print(f"\n‚úÖ APK trovato: {apk_path}")
    print(f"üì¶ Dimensione: {os.path.getsize(apk_path) / (1024*1024):.1f} MB")
    print("\nüì• Download in corso...")
    files.download(apk_path)
    print("\nüéâ FATTO! L'APK √® stato scaricato sul tuo computer!")
    print("   Trasferiscilo sul telefono e installalo.")
else:
    print("\n‚ùå ERRORE: APK non trovato!")
    print("   Verifica che la cella 3 sia stata completata con successo.")
    print("   Cerca errori nell'output della cella 3.")
````
